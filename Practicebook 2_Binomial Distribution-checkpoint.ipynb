```python
# Bionomial Distribution
# Problem Statement
# A bank issues credit card to its customers under the scheme of Master card. Based on past data, it is known that 60% of the customers pay the bill on time. If a sample of 7 accounts is selected at random from current database, construct the binomial probability distribution of accounts paying on time.
```

```python
# This module contains a large number of probability distributions, summary and frequency statistics, correlation functions and statistical tests, masked statistics, kernel density estimation, quasi-Monte Carlo functionality, and more.
# SciPy Stats can generate discrete or continuous random numbers.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n = 7
```

```python
p = .6
```

```python
# arange() function is used to get evenly spaced values within a given interval. 

k = np.arange(0,8)
k
```

```python
# stats. binom module can be used to calculate a binomial distribution using the values of n and p.

binomial = stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial, 'o-')
plt.title('Binomial: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Sucesses')
plt.show()
```

```python
plt.plot(k,binomial, '*--')
plt.title('Binomial: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Sucesses')
plt.show()
```
