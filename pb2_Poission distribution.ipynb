```python
# Poisson Distribution 
#Problem statement
# Given that a bank has a footfall of 6 people per minute, find out the 
# a) probabilty distribution for upto 20 people per minute.
# b) probabilty of at least 4 people per minute.

# SciPy Stats can generate discrete or continuous random numbers. It also consists of many other functions to generate descriptive statistical values. 

#improting required packages
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# average rate, denoted with lambda in the formula for poission distribution

rate = 6
```

```python
# a numpy aray of different no. of sucesses ranging from 0 to 19 to construct a probabilty distribution.

n = np.arange(0,20)
```

```python
# calculating the distribution and storing the distribution of probabilties in an array.

#Probability Mass Function.
poisson = stats.poisson.pmf(n,rate)
```

```python
# printing the ptrobabilty distribution for different values of x

poisson
```

```python
# probabilty of x>=4, i.e. if we consider bank example, probabilty of atleast 4 visitors per minute.
# given an average of 6 visitors per minute.
 
1- (poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
#visualising the plot with pyplot

plt.plot (n,poisson, 'o-')
plt.title('Poisson:$\lambda$ = %i' %rate)
plt.xlabel('Number of Customers Arriving in a minute')
plt.ylabel('Probability of numbar of customers arriving in a minute')
plt.show()
```
